# Wins (Dependent Variable)

## Which side wins more?

```python
blue_side = df['hasWon'].mean()
red_side = 1 - blue_side
labels = ['Blue', 'Red']
sizes = [blue_side,  red_side]
colors = ['#63D1F4', '#EE6363']

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        colors=colors, shadow=True, startangle=140)
plt.title("Win Percentage by Side")
plt.show()
```

## Blue vs Red side (Objectives)

```python
categories = ['hasWon', 'firstBlood', 'firstTower',
       'firstBaron', 'firstDragon', 'firstInhibitor',
       'firstRiftHerald','dragonKillsBlue','baronKillsBlue',
       'towerKillsBlue','inhibitorKillsBlue','riftHeraldKillsBlue']

blue_percentage_data = {}

for category in categories:
    if category.find("Blue") == -1:
        blue_percent = df[category].mean()
        red_percent = 1 - blue_percent
    else:
        blue_total = df[category].sum()
        category_red = category.replace('Blue','Red')
        red_total = df[category_red].sum()
        total = (blue_total + red_total).sum()
    
        blue_percent = blue_total/total
        red_percent = red_total/total
    blue_percentage_data[category] = blue_percent

blue_over = {k:v for k,v in blue_percentage_data.items()}
red_over = {k:1-v for k,v in blue_over.items()}

y = range(len(blue_over))
plt.figure(figsize=(20,10))
barWidth = 0.9
# blue
plt.bar(y, list(blue_over.values()), color='#0EBFE9', edgecolor='black', width=barWidth)
# red
plt.bar(y, list(red_over.values()), bottom=list(blue_over.values()), color='#cc0000', edgecolor='black', width=barWidth)

# Custom x axis
plt.xticks(y, blue_over.keys())
plt.ylim((0,1))
plt.ylabel("Percentage Taken")
plt.title("Differences in sides for values over 1%")
# Show graphic
plt.show()
```

# Gold

```python
sns.palplot(sns.color_palette('summer', 20))
colors = sns.color_palette('summer', 20)
```

```python
#create gold dataframe with all goldEarned's, gameDuration's and hasWon's values..
melt_gold = df[['hasWon','gameDuration','goldEarnedBlue','goldEarnedRed']]

#all goldEarned on victory side
def funcWin(row):
    if(row['hasWon']==0):
        row['goldEarnedBlue']=row['goldEarnedRed']
    return row

#all goldEarned on defeated side
def funcLose(row):
    if(row['hasWon']==0):
        row['goldEarnedRed']=row['goldEarnedBlue']
    return row

melt_gold_win = melt_gold.apply(funcWin,axis = 1)
melt_gold_win = melt_gold_win.drop('goldEarnedRed', 1)
melt_gold_win['hasWon'] = 1
melt_gold_win = melt_gold_win.rename(columns={'goldEarnedBlue': 'goldEarned'})

melt_gold_lose = melt_gold.apply(funcLose,axis = 1)
melt_gold_lose = melt_gold_lose.drop('goldEarnedBlue', 1)
melt_gold_lose['hasWon'] = 0
melt_gold_lose = melt_gold_lose.rename(columns={'goldEarnedRed': 'goldEarned'})

final_gold = melt_gold_win.append(melt_gold_lose, ignore_index=True)
final_gold = final_gold.astype(int)

# Create pal dictionary using comprehension
pal = {win: colors[2] if win == 1 else colors[19] for win in final_gold.hasWon.unique()}

fig, ax = plt.subplots(3,1, figsize=(14,14))
fig.suptitle('Total Gold Value - Win', fontsize=24, fontweight='bold', x=0.06, y=1.025, horizontalalignment='left')
fig.subplots_adjust(top=0.85)

plt.subplot(4,1,1)
p1 = sns.boxplot(x='gameDuration', y='goldEarned', hue='hasWon', data = final_gold[final_gold['gameDuration']<20], palette=pal)
box = p1.artists[18] # artists allow you to access the box appearance properties of the box you want to highlight
box.set_facecolor(sns.xkcd_rgb['blue'])
box = p1.artists[19] # artists allow you to access the box appearance properties of the box you want to highlight
box.set_facecolor(sns.xkcd_rgb['red'])
plt.ylabel('Gold Value')

plt.subplot(4,1,2)
ax = sns.boxplot(x='gameDuration', y='goldEarned', hue='hasWon', data = final_gold[(final_gold['gameDuration']>=20) & ((final_gold['gameDuration'] <40))],palette = pal)
plt.ylabel('Gold Value')

plt.subplot(4,1,3)
sns.boxplot(x='gameDuration', y='goldEarned', hue='hasWon', data = final_gold[(final_gold['gameDuration']>=40) & ((final_gold['gameDuration']<60))], palette=pal)
plt.ylabel('Gold Value')


plt.tight_layout()
plt.show()
```

# Gold and Champion Level

## Blue Side

```python
#creating the dataframe
lvl_gold = df[['hasWon','champLevelBlue','champLevelRed','goldEarnedBlue','goldEarnedRed']]

def funcLvlGoldBlue(row):
    row['goldEarnedBlue']=row['goldEarnedBlue'] - row['goldEarnedRed']
    row['champLevelBlue']=row['champLevelBlue'] - row['champLevelRed']
    return row

lvl_gold_blue = lvl_gold.apply(funcLvlGoldBlue,axis = 1)
lvl_gold_blue.drop(['goldEarnedRed', 'champLevelRed'], axis=1, inplace=True)
lvl_gold_blue = lvl_gold_blue.rename(columns={'goldEarnedBlue': 'blueGoldDiff','champLevelBlue':'blueLevelDiff'})

plt.figure(figsize=(8,6))
sns.scatterplot(x='blueLevelDiff', y='blueGoldDiff', hue='hasWon', data=lvl_gold_blue)
```
